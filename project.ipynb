#### Exercise 16. Calculate the sample mean and variance of the variables: price, rooms, surface area and bathrooms (you must use the subset obtained in the question 14) (★★★)

Print both values for each variable

```python
import pandas as pd
import numpy as np
# this CSV file contains semicolons instead of comas as separator
df = pd.read_csv('assets/real_estate.csv', sep=';')

south_df = ["Fuenlabrada","Leganés","Getafe","Alcorcón"]
filtered_df = df[df['level5'].isin(south_df)]

priceM = filtered_df.price.mean()
priceV = filtered_df.price.var()
print('Price mean and variance', priceM, priceV)

SurM = filtered_df.surface.mean()
SurV = filtered_df.surface.var()
print('Surface mean and variance', SurM, SurV)

batM = filtered_df.bathrooms.mean()
batV = filtered_df.bathrooms.var()
print('Bathroom mean and variance', batM, batV)

roomM = filtered_df.rooms.mean()
roomV = filtered_df.rooms.var()
print('Rooms mean and variance', roomM, roomV)
```
